**Problem 1. Range and For loop**

(a) For loopÏôÄ rangeÎ•º Ïù¥Ïö©ÌïòÏó¨ 0Î∂ÄÌÑ∞ 100ÍπåÏßÄ Ï†ïÏàòÎßå Ï∂úÎ†•ÌïòÎäî ÏΩîÎìúÎ•º ÏûëÏÑ±ÌïòÏãúÏò§.

```python
for i in range(101):
  print(i)
```

(b) For loopÏôÄ rangeÎ•º ÏÇ¨Ïö©ÌïòÏó¨ 0Î∂ÄÌÑ∞ 100ÍπåÏßÄ Ï†ïÏàò Ï§ë 7Î°ú ÎÇòÎàÑÏñ¥ Îñ®Ïñ¥ÏßÄÎäî Ï†ïÏàòÎßå Ï∂úÎ†•ÌïòÎäî ÏΩîÎìúÎ•º ÏûëÏÑ±ÌïòÏãúÏò§.

(c) For loopÏôÄ rangeÎ•º ÏÇ¨Ïö©ÌïòÏó¨ 1Î∂ÄÌÑ∞ 100ÍπåÏßÄ Ï†ïÏàò Ï§ë 5Î°ú ÎÇòÎàÑÏñ¥ Îñ®Ïñ¥ÏßÄÏßÄÎßå 3ÏúºÎ°ú ÎÇòÎàÑÏñ¥

(d) Ï†ïÏàòÎ•º ÏûÖÎ†•Î∞õÏùÄ ÌõÑ 1Î∂ÄÌÑ∞ ÏûÖÎ†•Î∞õÏùÄ Ï†ïÏàòÍπåÏßÄ ÏßùÏàòÎì§Ïùò Ìï©ÏùÑ Íµ¨ÌïòÎäî ÏΩîÎìúÎ•º ÏûëÏÑ±ÌïòÏãúÏò§.

- ÏûÖÎ†•Í∞íÏúºÎ°ú 27ÏùÑ ÎÑ£ÏùÄ ÌõÑ Í≤∞Í≥ºÍ∞íÏùÑ Ï≤®Î∂ÄÌï† Í≤É

```python
num=int(input("Ï†ïÏàòÎ•º ÏûÖÎ†•ÌïòÏÑ∏Ïöî: "))

n_sum=0
for i in range(1, num):
  if(i % 2 == 0):
    n_sum+=i

print("1Î∂ÄÌÑ∞ {0}ÍπåÏßÄ Ï†ïÏàò Ï§ë ÏßùÏàòÏùò Ìï©: {1}".format(num,n_sum))
```

**Problem 2. While loops**

(a) While loopsÎ•º Ïù¥Ïö©ÌïòÏó¨ 0Î∂ÄÌÑ∞ 100ÍπåÏßÄ 6ÏúºÎ°ú ÎÇòÎàÑÏñ¥ Îñ®Ïñ¥ÏßÄÎäî Ï†ïÏàòÎ•º Ï∂úÎ†•ÌïòÎäî ÏΩîÎìúÎ•º ÏûëÏÑ±ÌïòÏãúÏò§.

```python
num=0
while (num <= 100):
    if (num % 6 == 0):
        print(num)
    num+=1
```

(b) While loopsÎ•º Ïù¥Ïö©ÌïòÏó¨ 5,7, 11Î°ú ÎÇòÎàÑÏñ¥ Îñ®Ïñ¥ÏßÄÎäî Ï†ïÏàòÎ•º 20Í∞ú Ï∂úÎ†•ÌïòÎäî ÏΩîÎìúÎ•º ÏûëÏÑ±ÌïòÏãúÏò§.

```python
num=0
count=0
while (count <= 20):
  if(num % 5 == 0)&(num % 7 == 0)&(num % 11 == 0):
    print(num)
    count+=1
  num+=1
```

**Problem 3. Fibonacci sequence**

(a) Fibonacci sequenceÏùÑ Ìï®ÏàòÎ°ú Ï†ïÏùòÌïú ÌõÑ Í∑∏ ÏΩîÎìúÎ•º ÏûëÏÑ±ÌïòÏãúÏò§. ÎòêÌïú, Ï†ïÏùòÌïú Ìï®ÏàòÎ•º Ïù¥Ïö©ÌïòÏó¨ $ùëé35$Î•º Íµ¨ÌïòÏãúÏò§.

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1)+fibonacci(n - 2)

n = int(input("nÏùÑ ÏûÖÎ†•ÌïòÏÑ∏Ïöî: "))
a = fibonacci(n)

print(f"F({n})={a}")
```

**Problem 7. Function 2 and modules**

(a) ÏûÖÎ†•Îêú listÏóêÏÑú ÌôÄÏàòÏù∏ ÏõêÏÜåÎßå ÎπºÎÇ¥Îäî Ìï®ÏàòÎ•º ÎßåÎìúÏãúÏò§.

- Ìï®Ïàò Ïù¥Î¶ÑÏùÄ ‚Äòmyodd‚ÄôÎ°ú ÏßÄÏ†ïÌï† Í≤É.
- forÎ¨∏Í≥º ifÎ¨∏ÏùÑ Ïù¥Ïö©Ìï† Í≤É

```python
def myodd(x):
  odd=[]
  for i in x:
    if(i % 2 != 0):
      odd.append(i)
  return odd
```

(b) (a)ÏóêÏÑú ÎßåÎì† Ìï®ÏàòÎ•º mymoduleÏù¥ÎùºÎäî Ïù¥Î¶ÑÏùò .pyÎ°ú Ï†ÄÏû•ÌïòÏãúÏò§.

```python
code="""
def myodd(x):
  odd=[]
  for i in x:
    if(i % 2 != 0):
      odd.append(i)
  return odd
"""
with open('mymodule.py','w') as f:
    f.write(code)
```
