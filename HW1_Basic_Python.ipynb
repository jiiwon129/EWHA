**Problem 1. Range and For loop**

(a) For loop와 range를 이용하여 0부터 100까지 정수만 출력하는 코드를 작성하시오.

```python
for i in range(101):
  print(i)
```

(b) For loop와 range를 사용하여 0부터 100까지 정수 중 7로 나누어 떨어지는 정수만 출력하는 코드를 작성하시오.

```python
for i in range(101):
  if (i % 7 == 0):
    print(i)
```

(c) For loop와 range를 사용하여 1부터 100까지 정수 중 5로 나누어 떨어지지만 3으로 나누어
떨어지지 않는 정수를 출력하는 코드를 작성하시오.

```python
for i in range(101):
  if (i % 5 == 0)&(i % 3 != 0):
      print(i)
```

(d) 정수를 입력받은 후 1부터 입력받은 정수까지 짝수들의 합을 구하는 코드를 작성하시오.

- 입력값으로 27을 넣은 후 결과값을 첨부할 것
- For문과 if문을 이용할 것

```python
num=int(input("정수를 입력하세요: "))

n_sum=0
for i in range(1, num):
  if(i % 2 == 0):
    n_sum+=i

print("1부터 {0}까지 정수 중 짝수의 합: {1}".format(num,n_sum))
```

**Problem 2. While loops**

(a) While loops를 이용하여 0부터 100까지 6으로 나누어 떨어지는 정수를 출력하는 코드를 작성하시오.

```python
num=0
while (num <= 100):
    if (num % 6 == 0):
        print(num)
    num+=1
```

(b) While loops를 이용하여 5,7, 11로 나누어 떨어지는 정수를 20개 출력하는 코드를 작성하시오.

```python
num=0
count=0
while (count <= 20):
  if(num % 5 == 0)&(num % 7 == 0)&(num % 11 == 0):
    print(num)
    count+=1
  num+=1
```

**Problem 3. Fibonacci sequence**

(a) Fibonacci sequence을 함수로 정의한 후 그 코드를 작성하시오. 또한, 정의한 함수를 이용하여 $𝑎35$를 구하시오.

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1)+fibonacci(n - 2)

n = int(input("n을 입력하세요: "))
a = fibonacci(n)

print(f"F({n})={a}")
```

**Problem 6. Function 1**

(a) 다음 조건을 만족하는 함수를 만드시오. 함수 이름은 my_max(x, y)로 정의하며, x와 y가
input으로 들어간 후 둘 중 큰 수를 return한다. Python의 내장 함수 max()를 사용하지 않으며, if문을 이용하여 함수를 만든 후 그 코드를 작성하고, 임의의 두 숫자를 넣어 그 결과를 작성하시오.

```python
def my_max(x, y):
  if (x > y):
    return x
  else:
    return y

my_max(5,7)
```

(b) 1부터 100까지 0.1 씩 증가하여 숫자를 생성한 후 함수 $𝑦=3x^2−x+2$의 plot을 그리시오.

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(1,100.1,0.1)
y=3*x**2-x+2

plt.figure(figsize=(12, 6))
plt.plot(x,y,label='$y = 3x^2 - x + 2$',color='b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of $y = 3x^2 - x + 2$')
plt.legend()
plt.show()
```

**Problem 7. Function 2 and modules**

(a) 입력된 list에서 홀수인 원소만 빼내는 함수를 만드시오.

- 함수 이름은 ‘myodd’로 지정할 것.
- for문과 if문을 이용할 것

```python
def myodd(x):
  odd=[]
  for i in x:
    if(i % 2 != 0):
      odd.append(i)
  return odd
```

(b) (a)에서 만든 함수를 mymodule이라는 이름의 .py로 저장하시오.

```python
code="""
def myodd(x):
  odd=[]
  for i in x:
    if(i % 2 != 0):
      odd.append(i)
  return odd
"""
with open('mymodule.py','w') as f:
    f.write(code)
```
